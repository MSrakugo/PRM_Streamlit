```python
# Pandas DataFrame のカラム名を変換する例
df = pd.read_excel('Xpel2_Before_Preprocessed_ver240707.xlsx', index_col=0, header=0)
df.rename(columns=column_name_map, inplace=True)
```

```python
# インデックスに重複がないか確認
def check_index_duplicates(dataframe):
    if dataframe.index.has_duplicates:
        print("インデックスに重複があります。")
        duplicate_indices = dataframe.index[dataframe.index.duplicated()].unique()
        print("重複したインデックス: ", duplicate_indices)
    else:
        print("インデックスに重複はありません。")

def check_dtypes(dataframe):
    float_columns = dataframe.select_dtypes(include=['float']).columns
    non_float_columns = dataframe.columns.difference(float_columns)
    
    if len(non_float_columns) > 0:
        print("以下の列はfloat型ではありません:")
        for col in non_float_columns:
            print(f"{col}: {dataframe[col].dtype}")
    else:
        print("すべての列がfloat型です。")


# DataFrameのインデックスに重複がないか確認する
check_index_duplicates(df)

# DataFrameの各列がfloat型かどうか確認する
check_dtypes(df)
```
