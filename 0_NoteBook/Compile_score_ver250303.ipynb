```python
import pandas as pd
import glob
```

```python
path_list = glob.glob(glob.glob('*/')[1]+'*/')
data = pd.DataFrame()

for path in path_list:
    try:
        data_path = pd.read_excel(path+"Score_all.xlsx", index_col = 0).iloc[:, 0]
        data = pd.concat([data, data_path], axis = 1)
    except:
        print(path)
```
